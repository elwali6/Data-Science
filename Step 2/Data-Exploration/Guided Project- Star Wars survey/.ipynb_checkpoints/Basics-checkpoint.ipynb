```python
import pandas as pd

star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
#star_wars.head(10)
#star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars.head()
```

```python
yes_no = {'Yes': True, 'No': False}

star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()

column1 = 'Have you seen any of the 6 films in the Star Wars franchise?'
star_wars[column1] = star_wars[column1].map(yes_no)

column2 = 'Do you consider yourself to be a fan of the Star Wars film franchise?'
star_wars[column2] = star_wars[column2].map(yes_no)

star_wars.head(4)
```

```python
map_column3 = {"Star Wars: Episode I  The Phantom Menace": True,
    NaN: False}
map_column4 = {"Star Wars: Episode II Attack of the Clones": True,
    NaN: False}
map_column5 = {"Star Wars: Episode III Revenge of the Sith": True,
    NaN: False}
map_column6 = {"Star Wars: Episode IV A New Hope": True,
    NaN: False}
map_column7 = {"Star Wars: Episode V The Empire Strikes Back": True,
    NaN: False}
map_column8 = {"Star Wars: Episode VI Return of the Jedi": True,
    NaN: False}

all_columns_movies = star_wars.columns[3:9]

star_wars[all_columns_movies[0]] = star_wars[all_columns_movies[0]].map(map_column3)
star_wars.head()
```
