```python
import pandas as pd
import numpy as np

star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
#star_wars.head(10)
#star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars.head()
```

```python
yes_no = {'Yes': True, 'No': False}

star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()

column1 = 'Have you seen any of the 6 films in the Star Wars franchise?'
star_wars[column1] = star_wars[column1].map(yes_no)

column2 = 'Do you consider yourself to be a fan of the Star Wars film franchise?'
star_wars[column2] = star_wars[column2].map(yes_no)

star_wars.head(4)
```

```python
map_column = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}
    

for c in star_wars.columns[3:9]:
    star_wars[c] = star_wars[c].map(map_column)

star_wars.head()
```

```python
star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6",
})

star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars = star_wars.rename(columns={
        'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'ranking_1',
        'Unnamed: 10': 'ranking_2',
        'Unnamed: 11': 'ranking_3',
        'Unnamed: 12': 'ranking_4',
        'Unnamed: 13': 'ranking_5',
        'Unnamed: 14': 'ranking_6',
        
    })
star_wars.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

df = star_wars[star_wars.columns[9:15]]
mean_rankings = df.mean()
mean_rankings.plot(kind='bar')
plt.show()
```

```python
seen_df = star_wars[star_wars.columns[3:9]]
sum_seen = seen_df.sum()
sum_seen.plot(kind='bar')
plt.show()
```

```python
males = star_wars[star_wars['Gender'] == 'Male']
female = star_wars[star_wars['Gender'] == 'Female']

df = males[males.columns[9:15]]
mean_rankings_male = df.mean()
mean_rankings_male.plot(kind='bar')
plt.title('Mean Ranking Male')
plt.show()

seen_df = males[males.columns[3:9]]
sum_seen_male = seen_df.sum()
sum_seen_male.plot(kind='bar')
plt.title('Seen Ranking Male')
plt.show()

df = female[males.columns[9:15]]
mean_rankings_female = df.mean()
mean_rankings_female.plot(kind='bar')
plt.title('Mean Ranking Female')
plt.show()

seen_df = female[males.columns[3:9]]
sum_seen_female = seen_df.sum()
sum_seen_female.plot(kind='bar')
plt.title('Seen Ranking Female')
plt.show()



```
